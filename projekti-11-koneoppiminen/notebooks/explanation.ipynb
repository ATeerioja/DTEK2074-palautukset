## 0 Ongelman rajaaminen ja työkalujen valinta

Nopean tutkimuksen perusteella tunteiden luokitteluun on olemasssa rakenne. Käytän tässä työssä Paul Ekmanin tunnistamia tunteiden luokkia, joita on 7. Iloinen, surullinen, vihainen, yllättynyt, peloissaan, inhoontunut ja neutraali.

Koulutusdatan generoin itse omilla kasvoillani ja tunteillani. Koulutan myös toisen mallin julkisella aineistolla. Näiden kahden mallin tunteiden tunnistuskykyä voidaan verrata keskenään tuntemattomilla kasvoilla, josta voidaan havaita onko omalla datalla koulutettu malli hyvä yleistämään verrattuna isompaan ja monipuoliseen koulutusdataan.

Käytän ainakin seuraavia työkaluja projektin teossa.
- Python 3.10+
- OpenCV kasvojentunnistamiseen ja kuvaamiseen
- MediaPipe tai dlib kasvojen piirteiden tunnistamiseen
- PyTorch mallin kouluttamiseen
- Label Studio koulutusdatan annotointiin



## 1 Datan luonti ja luokittelu

### 1.1 Kuvien kerääminen
Otetaan kuvia verkkokameralla muutaman kerran sekunnissa, jotta saadaan kerättyä raakaa dataa. Otsikoidaan nämä kuvat automaattisesti, niin että kerrotaan ensin tunne ja sen jälkeen kuvataan tunnetta. Toistetaan tämä tekniikka kaikille tunteille. Tässä hieman naiivissa toteutuksessa mallille korostuu ylimääräinen informaatio datasta. Pyritään siis kuvaamaan hieman eri valaistuksella ja eri värisellä taustalla, jotta saadaan tätä virhettä minimoitua.

### 1.2 Kasvojen tunnistaminen
Käytetään MediaPipe työkalua kasvojen tunnistamiseen ja prosessointiin. Prosessoinnissa kasvojen alue rajataan laatikolla ja kasvot suunnataan silmien ja kasvojen piirteiden avulla samansuuntaisiksi, jotta kasvot ovat toisiinsa nähden samassa sijainnissa. Näin malli oppii vain kasvojen piirteiden tunnistamista, eikä tulkitse kasvojen sijaintia osaksi eri ilmeitä.

### 1.3 Kuvien otsikointi
Tarkistetaan kuvat Label Studio työkalulla, jotta turhat kuvat voidaan poistaa koulutusdatasta. Jos yksi luokka sisältää paljon virheellistä dataa, mallin on vaikeampi oppia tunnistamaan tunteiden ominaisia piireitä, joka tekee mallin yleistämisestä vaikeaa.

## 2 datan esikäsittely ja ominaisuusvektorin luonti

1. Pienennetään kuvan resoluutiota, jotta mallin kouluttamisessa ei kestä liian kauaa.
2. Muutetaan kuva mustavalkoiseksi, jotta mallin kouluttamisessa ei kestä liian kauaa. Mustavalkoinen voi myös parantaa koulutustuloksia pienellä otoskoolla, koska malli ei silloin keskity esimerkiksi ihonväriin tai taustan väriin yhtä paljon.
3. Normalisoidaan kuvan arvot välille [0 - 1], jotta mallin koulutus tapahtuu hitaammin ja kontrolloidummin.
4. Jaetaan data koulutus, validointi ja testiryhmiin. Tehdän jako stratifioidusti, jotta kaikkiin ryhmiin saadaan esimerkkejä jokaisesta luokasta.
